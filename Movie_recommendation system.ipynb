```python
def content_based_rec(title,no_of_rec=10):
    stri=" "
    ans=sig[df[title]]
    ans_list=sorted(list(enumerate(list(ans))),key=lambda x:x[1],reverse=True)
    ans_list=ans_list[1:no_of_rec]
    print("The recommendations are as follows:")
    for i in ans_list:
        stri=stri+" "+data['original_title'].iloc[i[0]]+" and \n"
    return stri
        
```

```python
content_based_rec('World War Z')
```
